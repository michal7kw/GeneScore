# Derive TF-target genes 

```python
import scanpy as sc
import mudata as md
from scenicplus.grn_builder.modules import merge_emodules
import pandas as pd
import numpy as np
```

```python
# Get the direct and extended eRegulon metadata
direct_eregulons = mdata.uns['direct_e_regulon_metadata']
extended_eregulons = mdata.uns['extended_e_regulon_metadata']
```

```python
extended_eregulons.head()
```

```python
# Function to process eRegulons
def process_eregulons(eregulons):
    data = []
    for _, row in eregulons.iterrows():
        data.append({
            'source': row['TF'],
            'target': row['Gene'],
            'coef_mean': row['rho_TF2G'],
            'coef_abs': abs(row['rho_TF2G']),
            'p': np.nan,  
            '-logp': 0.0, 
            'is_extended': row['is_extended'],
            'importance_TF2G': row['importance_TF2G'],
            'regulation': row['regulation'],
            'triplet_rank': row['triplet_rank']
        })
    return pd.DataFrame(data)
```

```python
# Process direct and extended eRegulons
direct_df = process_eregulons(direct_eregulons)
extended_df = process_eregulons(extended_eregulons)
```

```python
# Combine direct and extended
combined_df = pd.concat([direct_df, extended_df], ignore_index=True)
```

```python
# Sort the DataFrame
combined_df = combined_df.sort_values(['source', 'target'])
```

```python
combined_df.head()
```

```python
# Select and reorder columns
final_df = combined_df[['source', 'target', 'coef_mean', 'coef_abs', 'p', '-logp', 'is_extended', 'importance_TF2G', 'regulation', 'triplet_rank']]
```

```python
# Save to CSV
final_df.to_csv(os.path.join(scplus_path, 'tf_target_genes.csv'), index=False, sep='\t')
```

```python
# Print some statistics
print(f"Total number of TF-target interactions: {len(final_df)}")
print(f"Number of unique TFs: {final_df['source'].nunique()}")
print(f"Number of unique target genes: {final_df['target'].nunique()}")
print(f"Number of direct interactions: {len(final_df[~final_df['is_extended']])}")
print(f"Number of extended interactions: {len(final_df[final_df['is_extended']])}")
```
